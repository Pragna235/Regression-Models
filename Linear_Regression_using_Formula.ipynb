```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#plt.rcParams['figure.figsize'] = (20.0,10.0)

#Reading Data
data=pd.read_csv('https://www.statlearning.com/s/Advertising.csv')
print(data.shape)
data.head()
```

```python
#Collecting X and Y
X = data['TV'].values
Y = data['sales'].values
#print(X)
#print(Y)
```

m = sum of ((x-mean_x)*(y-mean_y)) / sum of ((x-mean_x)*^2)

```python
# Mean X and Y (Least Squares Method)
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
m=len(X)

# Using the formula to calculate slope(m) and intercept(c)
numerator = 0
denominator = 0
for i in range(m): #Least Squares Method 
#where, formula for m = sigma((x-mean_x)*(y-mean_y)) / sigma((x-mean_x)**2)
  numerator += (X[i] - mean_x) * (Y[i] - mean_y)
  denominator += (X[i] -mean_x) ** 2
slope = numerator / denominator
#Now, in y = mx+c, y=mean_y, x=mean_x, m=slope. So, c = y-mx
intercept = mean_y - (slope * mean_x)

print("Slope m = ",slope)
print("Intercept c = ",intercept)
```

```python
# Plotting Values and Regression Line
max_x = np.max(X) + 100
min_x = np.min(X) - 100

#Calculating Live Vaues x and y
x = np.linspace(min_x,max_x,1000)
y = slope*x + intercept 

# Plotting line (Predicted Values)
plt.plot(x,y,color='#58b970', label='Regression Line')
#Plotting Scatter Points (Actual Values)
plt.scatter(X,Y,c='#ef5423',label = "Scatter Plot")

plt.xlabel("TV")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
#r2 method for Regression (for efficiency)
ss_t = 0 #total sum of squares
ss_r = 0 #total sum of square of residuals
for i in range(m):
  y_pred = intercept + slope*X[i]
  ss_r += (Y[i] - y_pred) ** 2 
  ss_t += (Y[i] - mean_y) ** 2
r2 = 1-(ss_r/ss_t) #r2 formula = 1 - (RSS/TSS)
print(r2) #r2 should be greater than 0.6 for high efficiency
#Here, the efficiency is 0.6118750508500708, which is better compared to radio and newspaper
```
